```python
import kagglehub
path = kagglehub.dataset_download("mashlyn/online-retail-ii-uci")
```

```python
import pandas as pd
import os

print("Libraries pandas and os imported successfully.")

# List the contents of the path directory
print(f"Contents of the directory '{path}':")
print(os.listdir(path))
```

```python
data_file = os.path.join(path, 'online_retail_II.csv')
df = pd.read_csv(data_file)
print("Data loaded successfully. Displaying the first 5 rows:")
print(df.head())

print("\nInitial DataFrame Info:")
print(df.info())

# Handle missing values for CustomerID
initial_rows = df.shape[0]
df.dropna(subset=['CustomerID'], inplace=True)
print(f"\nRemoved {initial_rows - df.shape[0]} rows with missing CustomerID.")

# Remove cancelled orders (InvoiceNo starting with 'C')
df = df[~df['Invoice'].astype(str).str.startswith('C')]
print("Removed cancelled orders.")

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print("InvoiceDate converted to datetime.")

# Convert Quantity and UnitPrice to numeric, handling errors
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')

# Drop rows where Quantity or UnitPrice became NaN after coercion (if any non-numeric values existed)
df.dropna(subset=['Quantity', 'UnitPrice'], inplace=True)
print("Quantity and UnitPrice converted to numeric and rows with non-numeric values dropped.")

# Ensure Quantity and UnitPrice are positive
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print("Removed rows with non-positive Quantity or UnitPrice.")

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print("TotalPrice column created.")

print("\nCleaned DataFrame Info:")
print(df.info())
print("\nCleaned DataFrame head:")
print(df.head())
```

```python
data_file = os.path.join(path, 'online_retail_II.csv')
df = pd.read_csv(data_file)
print("Data loaded successfully. Displaying the first 5 rows:")
print(df.head())

print("\nInitial DataFrame Info:")
print(df.info())

# Handle missing values for CustomerID
initial_rows = df.shape[0]
df.dropna(subset=['Customer ID'], inplace=True) # Corrected column name
print(f"\nRemoved {initial_rows - df.shape[0]} rows with missing Customer ID.")

# Remove cancelled orders (InvoiceNo starting with 'C')
df = df[~df['Invoice'].astype(str).str.startswith('C')]
print("Removed cancelled orders.")

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print("InvoiceDate converted to datetime.")

# Convert Quantity and Price to numeric, handling errors
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce') # Corrected column name

# Drop rows where Quantity or Price became NaN after coercion (if any non-numeric values existed)
df.dropna(subset=['Quantity', 'Price'], inplace=True) # Corrected column name
print("Quantity and Price converted to numeric and rows with non-numeric values dropped.")

# Ensure Quantity and Price are positive
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)] # Corrected column name
print("Removed rows with non-positive Quantity or Price.")

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['Price'] # Corrected column name
print("TotalPrice column created.")

print("\nCleaned DataFrame Info:")
print(df.info())
print("\nCleaned DataFrame head:")
print(df.head())
```

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"Snapshot Date for RFM analysis: {snapshot_date}")

# Calculate RFM features
rfm_df = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('Invoice', 'nunique'),
    MonetaryValue=('TotalPrice', 'sum')
).reset_index()

# Calculate Average Order Value (AOV)
rfm_df['AOV'] = rfm_df['MonetaryValue'] / rfm_df['Frequency']

print("\nRFM DataFrame Head:")
print(rfm_df.head())

print("\nRFM DataFrame Info:")
print(rfm_df.info())
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb

# Define target and features
X = rfm_df[['Recency', 'Frequency', 'AOV']]
y = rfm_df['MonetaryValue']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data split into training and testing sets successfully.")
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# Instantiate and train XGBRegressor model
xgb_model = xgb.XGBRegressor(random_state=42)
xgb_model.fit(X_train, y_train)

print("\nXGBoost model trained successfully:")
print(xgb_model)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Make predictions on the test set
y_pred = xgb_model.predict(X_test)

# Evaluate model performance
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\nModel Evaluation on Test Set:")
print(f"R-squared: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Create a DataFrame for segmentation including Customer ID and predicted LTV
# Ensure proper alignment of Customer IDs from the test set with the predictions
segmentation_df = pd.DataFrame({
    'Customer ID': rfm_df.loc[X_test.index, 'Customer ID'],
    'PredictedLTV': y_pred
}).reset_index(drop=True)

print("Segmentation DataFrame created with Customer ID and PredictedLTV:")
print(segmentation_df.head())
print(segmentation_df.info())
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
segmentation_df['Cluster'] = kmeans.fit_predict(segmentation_df[['PredictedLTV']])

print("Segmentation DataFrame with Cluster labels:")
print(segmentation_df.head())

# Calculate the mean PredictedLTV for each cluster
cluster_means = segmentation_df.groupby('Cluster')['PredictedLTV'].mean().sort_values(ascending=False)
print("\nMean PredictedLTV for each cluster:")
print(cluster_means)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Visualize Model's Performance (Actual vs. Predicted LTV)
plt.figure(figsize=(10, 7))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect Prediction')
plt.xlabel('Actual LTV')
plt.ylabel('Predicted LTV')
plt.title('Actual vs. Predicted LTV')
plt.legend()
plt.grid(True)
plt.show()

# 2. Visualize Feature Importance
feature_importances = pd.Series(xgb_model.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.index, y=feature_importances.values)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('XGBoost Feature Importance for LTV Prediction')
plt.grid(axis='y')
plt.show()

# 3. Visualize Distribution of Customer Segments based on Predicted LTV
plt.figure(figsize=(12, 7))
sns.histplot(data=segmentation_df, x='PredictedLTV', hue='Cluster', kde=True, palette='viridis', multiple='stack')
plt.xlabel('Predicted LTV')
plt.ylabel('Count')
plt.title('Distribution of Predicted LTV by Customer Segment')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

print("Visualizations for model performance, feature importance, and customer segmentation generated successfully.")
```
